```python
import pandas as pd
import numpy as np
import os
import requests
```

# API DATA

```python
item_list = []

url = "https://datadiscovery.nlm.nih.gov/resource/crzr-uvwg.json"
response = requests.get(url).json()

for item in response:
    # Extract the necessary fields from the JSON and create a dictionary
    item_dict = {
      'id': item['id'],
      'enabled': item['enabled'],
      'created_at': item['created_at'],
      'updated_at': item['updated_at'],
      'spp': item['spp'],
      'setid': item['setid'],
      'splsize': item['splsize'],
      'splshape': item['splshape'],
      'splshape_text': item['splshape_text'],
      'splcolor': item['splcolor'],
      'splcolor_text': item['splcolor_text'],
      'source': item['source'],
      'product_code': item['product_code'],
      'part_num': item['part_num'],
      'ndc9': item['ndc9'],
      'ndc_labeler_code': item['ndc_labeler_code'],
      'ndc_product_code': item['ndc_product_code'],
      'medicine_name': item['medicine_name'],
      'effective_time': item['effective_time'],
      'file_name': item['file_name'],
      'dosage_form': item['dosage_form'],
      'author_type': item['author_type'],
      'author': item['author'],
      'has_image': item['has_image'],
      'version_number': item['version_number'],
      'updated': item['updated'],
      'stale': item['stale'],
      'new': item['new'],
      'pillbox_value': item['pillbox_value']}
    
    if 'dea_schedule_code' in item:
        item_dict['dea_schedule_code'] = item['dea_schedule_code']
        
    if 'dea_schedule_name' in item:
        item_dict['dea_schedule_name'] = item['dea_schedule_name']

    if 'rxnorm_update_time' in item:
        item_dict['rxnorm_update_time'] = item['rxnorm_update_time']

    if 'spl_inactive_ing' in item:
        item_dict['spl_inactive_ing'] = item['spl_inactive_ing']
        
    if 'spl_strength' in item:
        item_dict['spl_strength'] = item['spl_strength']

    if 'spl_ingredients' in item:
        item_dict['spl_ingredients'] = item['spl_ingredients']
        
    if 'splimprint' in item:
        item_dict['splimprint'] = item['splimprint']
        
    if 'marketing_act_code' in item:
        item_dict['marketing_act_code'] = item['marketing_act_code']
        
    if 'approval_code' in item:
        item_dict['approval_code'] = item['approval_code']
        
    if 'splscore' in item:
        item_dict['splscore'] = item['splscore']
    
    item_list.append(item_dict)
    
df = pd.DataFrame(item_list)
df
```

```python
df["api_code"] = df["ndc_labeler_code"]+"-"+df["ndc_product_code"]
```

```python
pillbox_limited = df[["ndc9","setid"]].drop_duplicates()
pillbox_limited
```

```python
item_list = []

for i in pillbox_limited["setid"]:
    
    url = f'https://api.fda.gov/drug/ndc.json?search={i}'
    response = requests.get(url).json()
    
    item_dict = {}
    
    item_dict['setid'] = i
    
    for item in response["results"]:
    
        if 'route' in item:
            item_dict['route'] = item['route'][0]

            item_list.append(item_dict)
            
df_api = pd.DataFrame(item_list)
df_api
```

```python
item_list = []

for i in pillbox_limited["setid"]:
    
    url = f'https://api.fda.gov/drug/label.json?search={i}'
    response = requests.get(url).json()
    
    item_dict = {}
    
    item_dict['setid'] = i
    
    for item in response["results"]:
    
        if 'spl_product_data_elements' in item:
            item_dict['spl_product_data_elements'] = item['spl_product_data_elements'][0]
            
        if 'spl_unclassified_section' in item:
            item_dict['spl_unclassified_section'] = item['spl_unclassified_section'][0]
            
        if 'boxed_warning' in item:
            item_dict['boxed_warning'] = item['boxed_warning'][0]
            
        if 'description' in item:
            item_dict['description'] = item['description'][0]
            
        if 'clinical_pharmacology' in item:
            item_dict['clinical_pharmacology'] = item['clinical_pharmacology'][0]
                      
        if 'indications_and_usage' in item:
            item_dict['indications_and_usage'] = item['indications_and_usage'][0]
            
        if 'contraindications' in item:
            item_dict['contraindications'] = item['contraindications'][0]
            
        if 'warnings' in item:
            item_dict['warnings'] = item['warnings'][0]
            
        if 'precautions' in item:
            item_dict['precautions'] = item['precautions'][0]
            
        if 'pregnancy' in item:
            item_dict['pregnancy'] = item['pregnancy'][0]
            
        if 'nursing_mothers' in item:
            item_dict['nursing_mothers'] = item['nursing_mothers'][0]
            
        if 'pediatric_use' in item:
            item_dict['pediatric_use'] = item['pediatric_use'][0]
            
        if 'overdosage' in item:
            item_dict['overdosage'] = item['overdosage'][0]
            
        if 'information_for_patients' in item:
            item_dict['information_for_patients'] = item['information_for_patients'][0]
            
        if 'adverse_reactions' in item:
            item_dict['adverse_reactions'] = item['adverse_reactions'][0]
            
        if 'dosage_and_administration' in item:
            item_dict['dosage_and_administration'] = item['dosage_and_administration'][0]   
            
        if 'how_supplied' in item:
            item_dict['how_supplied'] = item['how_supplied'][0]
            
            item_list.append(item_dict)
            
df_api2 = pd.DataFrame(item_list)
df_api2
```

```python
api_total = pd.merge(df_api,df_api2,on="setid",how="left")
pillbox_add = pd.merge(pillbox_limited,api_total,on="setid",how="left")
pillbox_add
```
